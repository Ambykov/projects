```python
# --------------------------
# –ö–∞—Ç–µ–≥–æ—Ä–∏–∏ —Ç–æ–≤–∞—Ä–æ–≤
# --------------------------

product_categories = {
    "–∫—É—Ö–æ–Ω–Ω–∞—è —Ç–µ—Ö–Ω–∏–∫–∞": ["–∫–æ—Ñ–µ–≤–∞—Ä–∫–∞", "—á–∞–π–Ω–∏–∫", "–º–∏–∫—Ä–æ–≤–æ–ª–Ω–æ–≤–∫–∞", "–º—É–ª—å—Ç–∏–≤–∞—Ä–∫–∞"],
    "–±—ã—Ç–æ–≤–∞—è —Ç–µ—Ö–Ω–∏–∫–∞": ["—Å—Ç–∏—Ä–∞–ª—å–Ω–∞—è –º–∞—à–∏–Ω–∞", "—É—Ç—é–≥", "—Ñ–µ–Ω", "–ø—ã–ª–µ—Å–æ—Å", "—Ö–æ–ª–æ–¥–∏–ª—å–Ω–∏–∫"],
    "–ø–æ—Å—É–¥–∞": ["—Å–∫–æ–≤–æ—Ä–æ–¥–∞", "–∫–∞—Å—Ç—Ä—é–ª—è", "–ø—Ä–æ—Ç–∏–≤–µ–Ω—å", "–∫—Ä—ã—à–∫–∞", "–Ω–æ–∂ –∫—É—Ö–æ–Ω–Ω—ã–π"],
    "–∞–∫—Å–µ—Å—Å—É–∞—Ä—ã –¥–ª—è —Ç–µ—Ö–Ω–∏–∫–∏": ["–∑–∞—Ä—è–¥–Ω–æ–µ —É—Å—Ç—Ä–æ–π—Å—Ç–≤–æ", "–∫–∞–±–µ–ª—å", "—Ñ–∏–ª—å—Ç—Ä –¥–ª—è –≤–æ–¥—ã", "–Ω–∞—Å–∞–¥–∫–∞ –¥–ª—è –ø—ã–ª–µ—Å–æ—Å–∞"],
    "–∫—Ä–∞—Å–æ—Ç–∞ –∏ –∑–¥–æ—Ä–æ–≤—å–µ": ["—ç–ø–∏–ª—è—Ç–æ—Ä", "–º–∞—à–∏–Ω–∫–∞ –¥–ª—è —Å—Ç—Ä–∏–∂–∫–∏", "–±—Ä–∏—Ç–≤–∞", "–≤—ã–ø—Ä—è–º–∏—Ç–µ–ª—å –≤–æ–ª–æ—Å"],
    "–¥–æ–º –∏ —Å–∞–¥": ["–ª–æ–ø–∞—Ç–∞", "—Å–µ–∫–∞—Ç–æ—Ä", "–ø–æ–ª–∏–≤–æ—á–Ω—ã–π —à–ª–∞–Ω–≥"],
    "—ç–ª–µ–∫—Ç—Ä–æ–Ω–∏–∫–∞": ["—Ç–µ–ª–µ–≤–∏–∑–æ—Ä", "—Å–º–∞—Ä—Ç—Ñ–æ–Ω", "–ø–ª–∞–Ω—à–µ—Ç", "–Ω–∞—É—à–Ω–∏–∫–∏"]
}

def get_random_product():
    category = choice(list(product_categories.keys()))
    product = choice(product_categories[category])
    return product, category
```

```python
# --------------------------
# –•–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏ —Ç–æ–≤–∞—Ä–æ–≤
# --------------------------

characteristics = {
    "—Ü–≤–µ—Ç": ["–∫—Ä–∞—Å–Ω—ã–π", "—Å–µ—Ä—ã–π", "–±–µ–∂–µ–≤—ã–π", "—á–µ—Ä–Ω—ã–π", "—Å–∏–Ω–∏–π", "–∑–æ–ª–æ—Ç–æ–π", "–º–∞—Ç–æ–≤—ã–π —á–µ—Ä–Ω—ã–π", "–≥–ª—è–Ω—Ü–µ–≤—ã–π –±–µ–ª—ã–π"],
    "—Ä–∞–∑–º–µ—Ä": ["51 —Å–º", "45 —Å–º", "S", "L", "XL", "M", "–±–æ–ª—å—à–æ–π", "–º–∞–ª–µ–Ω—å–∫–∏–π", "–æ—á–µ–Ω—å –º–∞–ª–µ–Ω—å–∫–∏–π"],
    "–º–∞—Ç–µ—Ä–∏–∞–ª": ["–ø–ª–∞—Å—Ç–∏–∫", "–º–µ—Ç–∞–ª–ª", "–∞–ª—é–º–∏–Ω–∏–π", "–Ω–µ—Ä–∂–∞–≤–µ—é—â–∞—è —Å—Ç–∞–ª—å", "—Å—Ç–µ–∫–ª–æ", "—Å–∏–ª–∏–∫–æ–Ω"],
    "–º–æ–¥–µ–ª—å": ["5 —Å–µ—Ä–∏–∏", "–º–æ–¥–µ–ª—å X", "–≤–µ—Ä—Å–∏—è 2022", "Pro-–º–æ–¥–µ–ª—å", "—Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–∞—è –º–æ–¥–µ–ª—å", "—É–ª—å—Ç—Ä–∞"],
    "–±—Ä–µ–Ω–¥": ["Samsung", "Bosch", "Tefal", "Dyson", "Redmond", "Philips", "Xiaomi", "Braun"]
}

def get_characteristic():
    color = random.choice(characteristics["—Ü–≤–µ—Ç"])
    size = random.choice(characteristics["—Ä–∞–∑–º–µ—Ä"])
    material = random.choice(characteristics["–º–∞—Ç–µ—Ä–∏–∞–ª"])
    model = random.choice(characteristics["–º–æ–¥–µ–ª—å"])
    brand = random.choice(characteristics["–±—Ä–µ–Ω–¥"])

    return {
        "—Ü–≤–µ—Ç": color,
        "—Ä–∞–∑–º–µ—Ä": size,
        "–º–∞—Ç–µ—Ä–∏–∞–ª": material,
        "–º–æ–¥–µ–ª—å": model,
        "–±—Ä–µ–Ω–¥": brand
    }
```

```python
# --------------------------
# –§—É–Ω–∫—Ü–∏—è –ø–∞—Ñ—Ä–∞–∑–æ–≤
# --------------------------

synonyms = {
    '–ø—Ä–∏—Å–ª–∞–ª–∏ –¥—Ä—É–≥–æ–π —Ç–æ–≤–∞—Ä': ['–ø—Ä–∏—à–ª–∞ –¥—Ä—É–≥–∞—è –º–æ–¥–µ–ª—å', '–ø–æ–ª—É—á–∏–ª(–∞) –Ω–µ —Ç–æ—Ç —Ç–æ–≤–∞—Ä', '–¥—Ä—É–≥–æ–π —Ü–≤–µ—Ç –∏ —Ä–∞–∑–º–µ—Ä'],
    '–Ω–µ —Ä–∞–±–æ—Ç–∞–µ—Ç': ['—Å–ª–æ–º–∞–ª—Å—è', '–Ω–µ –≤–∫–ª—é—á–∞–µ—Ç—Å—è', '–Ω–µ —Ñ—É–Ω–∫—Ü–∏–æ–Ω–∏—Ä—É–µ—Ç'],
    '–Ω–µ –æ—Ä–∏–≥–∏–Ω–∞–ª': ['–ø–æ–¥–¥–µ–ª–∫–∞', '–∫–æ–Ω—Ç—Ä–∞—Ñ–∞–∫—Ç', '–Ω–µ –Ω–∞—Å—Ç–æ—è—â–∏–π –±—Ä–µ–Ω–¥', '—Ñ–µ–π–∫']
}

def paraphrase_phrase(phrase):
    return random.choice(synonyms.get(phrase, [phrase]))
```

```python
# --------------------------
# –§—É–Ω–∫—Ü–∏–∏ generate_review –∏ generate_review_double
# --------------------------

def generate_review(phrase, phrase_type, product):
    # –í–∞—à–∞ —Ç–µ–∫—É—â–∞—è —Ä–µ–∞–ª–∏–∑–∞—Ü–∏—è
    chars = get_characteristic()
    if phrase_type in templates_by_type:
        template = random.choice(templates_by_type[phrase_type])
    else:
        template = "{start} {phrase} ‚Äî –±–æ–ª—å—à–µ –Ω–∏–∫–æ–≥–¥–∞ –Ω–µ –∑–∞–∫–∞–∂—É –∑–¥–µ—Å—å. {end}"

    try:
        return template.format(
            start=random.choice(emotion_starts),
            end=random.choice(emotion_ends),
            phrase=phrase,
            product=product,
            color=chars['—Ü–≤–µ—Ç'],
            size=chars['—Ä–∞–∑–º–µ—Ä'],
            material=chars['–º–∞—Ç–µ—Ä–∏–∞–ª'],
            brand=chars['–±—Ä–µ–Ω–¥']
        ).capitalize() + ('!' if random.random() > 0.7 else '.')
    except KeyError:
        return template.replace("{phrase}", phrase).replace("{product}", product)

def generate_review_double(phrase1, phrase2, product):
    chars = get_characteristic()

    # –®–∞–±–ª–æ–Ω—ã –¥–ª—è –¥–≤—É—Ö —Ñ—Ä–∞–∑
    template = random.choice([
        "{start} {phrase1} –∏ {phrase2} ‚Äî –±–æ–ª—å—à–µ –Ω–∏–∫–æ–≥–¥–∞ –Ω–µ –∑–∞–∫–∞–∂—É –∑–¥–µ—Å—å. {end}",
        "{start} –°–Ω–∞—á–∞–ª–∞ {phrase1}, –ø–æ—Ç–æ–º {phrase2} ‚Äî —Ç–æ–≤–∞—Ä –±–µ—Å–ø–æ–ª–µ–∑–µ–Ω. {end}",
        "{start} {phrase1}, –∞ –µ—â—ë {phrase2} ‚Äî –∑–∞—Å—Ç–∞–≤–∏–ª(–∞) –ø–µ—Ä–µ—Å–º–∞—Ç—Ä–∏–≤–∞—Ç—å –≤—ã–±–æ—Ä –º–∞–≥–∞–∑–∏–Ω–∞. {end}"
    ])

    try:
        return template.format(
            start=random.choice(emotion_starts),
            end=random.choice(emotion_ends),
            phrase1=phrase1,
            phrase2=phrase2,
            product=product,
            color=chars['—Ü–≤–µ—Ç'],
            material=chars['–º–∞—Ç–µ—Ä–∏–∞–ª'],
            brand=chars['–±—Ä–µ–Ω–¥']
        ).capitalize() + ('!' if random.random() > 0.7 else '.')
    except KeyError:
        return template.replace("{phrase1}", phrase1).replace("{phrase2}", phrase2)
```

```python
# --------------------------
# –ì–µ–Ω–µ—Ä–∞—Ü–∏—è –æ–¥–∏–Ω–æ—á–Ω—ã—Ö –æ—Ç–∑—ã–≤–æ–≤
# --------------------------

single_results = []

for group in key_words_all:
    code = group["code"]
    phrase_type = group["type"]
    phrases = group["phrases"]

    for phrase in phrases:
        for _ in range(5):  # –ø–æ 5 –≤–∞—Ä–∏–∞–Ω—Ç–æ–≤ –Ω–∞ –∫–∞–∂–¥—É—é —Ñ—Ä–∞–∑—É
            product, category = get_random_product()
            review = generate_review(phrase, phrase_type, product)

            single_results.append({
                "–¢–∏–ø –æ—Ç–∑—ã–≤–∞": "–û–¥–∏–Ω–æ—á–Ω–∞—è —Ñ—Ä–∞–∑–∞",
                "–ö–æ–¥": code,
                "–§—Ä–∞–∑–∞ 1": phrase,
                "–§—Ä–∞–∑–∞ 2": None,
                "–¢–æ–≤–∞—Ä": product,
                "–ö–∞—Ç–µ–≥–æ—Ä–∏—è —Ç–æ–≤–∞—Ä–∞": category,
                "–¶–≤–µ—Ç": get_characteristic()["—Ü–≤–µ—Ç"],
                "–ú–∞—Ç–µ—Ä–∏–∞–ª": get_characteristic()["–º–∞—Ç–µ—Ä–∏–∞–ª"],
                "–ë—Ä–µ–Ω–¥": get_characteristic()["–±—Ä–µ–Ω–¥"],
                "–û—Ç–∑—ã–≤": review
            })

print(f"‚úÖ –°–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞–Ω–æ {len(single_results)} –æ—Ç–∑—ã–≤–æ–≤ –ø–æ –æ–¥–Ω–æ–π —Ñ—Ä–∞–∑–µ")



# --------------------------
# –ì–µ–Ω–µ—Ä–∞—Ü–∏—è –æ—Ç–∑—ã–≤–æ–≤ —Å –¥–≤—É–º—è —Ñ—Ä–∞–∑–∞–º–∏ (–ø–æ code)
# --------------------------

pair_results = []

# –õ–∏–º–∏—Ç—ã –ø–∞—Ä
pair_limits = {
    102: 600,
    800: 100,
    801: 600
}

# –ì—Ä—É–ø–ø–∏—Ä—É–µ–º —Ñ—Ä–∞–∑—ã –ø–æ code
code_to_phrases = defaultdict(list)
for group in key_words_all:
    code = group["code"]
    code_to_phrases[code].extend(group["phrases"])

# –£–±–∏—Ä–∞–µ–º –¥—É–±–ª–∏–∫–∞—Ç—ã
for code in code_to_phrases:
    code_to_phrases[code] = list(set(code_to_phrases[code]))

# –ì–µ–Ω–µ—Ä–∏—Ä—É–µ–º –ø–∞—Ä–Ω—ã–µ –æ—Ç–∑—ã–≤—ã
for code, all_phrases in code_to_phrases.items():
    total_phrases = len(all_phrases)

    # –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–∞—Ä
    total_pairs = total_phrases * (total_phrases - 1) // 2
    max_by_ratio = round(total_pairs * 2 / 3)
    max_allowed = pair_limits.get(code, 100)
    num_pairs = min(max_by_ratio, max_allowed)

    print(f"–ö–æ–¥ {code}, –≤—Å–µ–≥–æ —Ñ—Ä–∞–∑: {total_phrases} ‚Üí –±—É–¥–µ–º –≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞—Ç—å –ø–∞—Ä: {num_pairs}")

    generated_pairs = set()

    while len(generated_pairs) < num_pairs and len(generated_pairs) < total_pairs:
        phrase1 = random.choice(all_phrases)
        phrase2 = random.choice([p for p in all_phrases if p != phrase1])

        pair = tuple(sorted((phrase1, phrase2)))

        if pair not in generated_pairs:
            generated_pairs.add(pair)

            product, category = get_random_product()
            review = generate_review_double(phrase1, phrase2, product)

            pair_results.append({
                "–¢–∏–ø –æ—Ç–∑—ã–≤–∞": "–î–≤–µ —Ñ—Ä–∞–∑—ã",
                "–ö–æ–¥": code,
                "–§—Ä–∞–∑–∞ 1": phrase1,
                "–§—Ä–∞–∑–∞ 2": phrase2,
                "–¢–æ–≤–∞—Ä": product,
                "–ö–∞—Ç–µ–≥–æ—Ä–∏—è —Ç–æ–≤–∞—Ä–∞": category,
                "–¶–≤–µ—Ç": get_characteristic()["—Ü–≤–µ—Ç"],
                "–ú–∞—Ç–µ—Ä–∏–∞–ª": get_characteristic()["–º–∞—Ç–µ—Ä–∏–∞–ª"],
                "–ë—Ä–µ–Ω–¥": get_characteristic()["–±—Ä–µ–Ω–¥"],
                "–û—Ç–∑—ã–≤": review
            })

print(f"‚úÖ –°–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞–Ω–æ {len(pair_results)} –æ—Ç–∑—ã–≤–æ–≤ –ø–æ –¥–≤—É–º —Ñ—Ä–∞–∑–∞–º")



# --------------------------
# –û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
# --------------------------

all_reviews = single_results + pair_results

# –°–æ—Ö—Ä–∞–Ω—è–µ–º –≤ Excel
df = pd.DataFrame(all_reviews)
df.to_excel("—É–ª—É—á—à–µ–Ω–Ω—ã–µ_–Ω–µ–≥–∞—Ç–∏–≤–Ω—ã–µ_–æ—Ç–∑—ã–≤—ã.xlsx", index=False)

print(f"\n‚úÖ –í—Å–µ–≥–æ —Å–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞–Ω–æ: {len(df)} –æ—Ç–∑—ã–≤–æ–≤")
print("üìå –†–µ–∑—É–ª—å—Ç–∞—Ç —Å–æ—Ö—Ä–∞–Ω—ë–Ω –≤ —Ñ–∞–π–ª: '—É–ª—É—á—à–µ–Ω–Ω—ã–µ_–Ω–µ–≥–∞—Ç–∏–≤–Ω—ã–µ_–æ—Ç–∑—ã–≤—ã.xlsx'")
```
