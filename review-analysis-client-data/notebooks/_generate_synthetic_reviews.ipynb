```python
# --------------------------
# Категории товаров
# --------------------------

product_categories = {
    "кухонная техника": ["кофеварка", "чайник", "микроволновка", "мультиварка"],
    "бытовая техника": ["стиральная машина", "утюг", "фен", "пылесос", "холодильник"],
    "посуда": ["сковорода", "кастрюля", "противень", "крышка", "нож кухонный"],
    "аксессуары для техники": ["зарядное устройство", "кабель", "фильтр для воды", "насадка для пылесоса"],
    "красота и здоровье": ["эпилятор", "машинка для стрижки", "бритва", "выпрямитель волос"],
    "дом и сад": ["лопата", "секатор", "поливочный шланг"],
    "электроника": ["телевизор", "смартфон", "планшет", "наушники"]
}

def get_random_product():
    category = choice(list(product_categories.keys()))
    product = choice(product_categories[category])
    return product, category
```

```python
# --------------------------
# Характеристики товаров
# --------------------------

characteristics = {
    "цвет": ["красный", "серый", "бежевый", "черный", "синий", "золотой", "матовый черный", "глянцевый белый"],
    "размер": ["51 см", "45 см", "S", "L", "XL", "M", "большой", "маленький", "очень маленький"],
    "материал": ["пластик", "металл", "алюминий", "нержавеющая сталь", "стекло", "силикон"],
    "модель": ["5 серии", "модель X", "версия 2022", "Pro-модель", "стандартная модель", "ультра"],
    "бренд": ["Samsung", "Bosch", "Tefal", "Dyson", "Redmond", "Philips", "Xiaomi", "Braun"]
}

def get_characteristic():
    color = random.choice(characteristics["цвет"])
    size = random.choice(characteristics["размер"])
    material = random.choice(characteristics["материал"])
    model = random.choice(characteristics["модель"])
    brand = random.choice(characteristics["бренд"])

    return {
        "цвет": color,
        "размер": size,
        "материал": material,
        "модель": model,
        "бренд": brand
    }
```

```python
# --------------------------
# Функция пафразов
# --------------------------

synonyms = {
    'прислали другой товар': ['пришла другая модель', 'получил(а) не тот товар', 'другой цвет и размер'],
    'не работает': ['сломался', 'не включается', 'не функционирует'],
    'не оригинал': ['подделка', 'контрафакт', 'не настоящий бренд', 'фейк']
}

def paraphrase_phrase(phrase):
    return random.choice(synonyms.get(phrase, [phrase]))
```

```python
# --------------------------
# Функции generate_review и generate_review_double
# --------------------------

def generate_review(phrase, phrase_type, product):
    # Ваша текущая реализация
    chars = get_characteristic()
    if phrase_type in templates_by_type:
        template = random.choice(templates_by_type[phrase_type])
    else:
        template = "{start} {phrase} — больше никогда не закажу здесь. {end}"

    try:
        return template.format(
            start=random.choice(emotion_starts),
            end=random.choice(emotion_ends),
            phrase=phrase,
            product=product,
            color=chars['цвет'],
            size=chars['размер'],
            material=chars['материал'],
            brand=chars['бренд']
        ).capitalize() + ('!' if random.random() > 0.7 else '.')
    except KeyError:
        return template.replace("{phrase}", phrase).replace("{product}", product)

def generate_review_double(phrase1, phrase2, product):
    chars = get_characteristic()

    # Шаблоны для двух фраз
    template = random.choice([
        "{start} {phrase1} и {phrase2} — больше никогда не закажу здесь. {end}",
        "{start} Сначала {phrase1}, потом {phrase2} — товар бесполезен. {end}",
        "{start} {phrase1}, а ещё {phrase2} — заставил(а) пересматривать выбор магазина. {end}"
    ])

    try:
        return template.format(
            start=random.choice(emotion_starts),
            end=random.choice(emotion_ends),
            phrase1=phrase1,
            phrase2=phrase2,
            product=product,
            color=chars['цвет'],
            material=chars['материал'],
            brand=chars['бренд']
        ).capitalize() + ('!' if random.random() > 0.7 else '.')
    except KeyError:
        return template.replace("{phrase1}", phrase1).replace("{phrase2}", phrase2)
```

```python
# --------------------------
# Генерация одиночных отзывов
# --------------------------

single_results = []

for group in key_words_all:
    code = group["code"]
    phrase_type = group["type"]
    phrases = group["phrases"]

    for phrase in phrases:
        for _ in range(5):  # по 5 вариантов на каждую фразу
            product, category = get_random_product()
            review = generate_review(phrase, phrase_type, product)

            single_results.append({
                "Тип отзыва": "Одиночная фраза",
                "Код": code,
                "Фраза 1": phrase,
                "Фраза 2": None,
                "Товар": product,
                "Категория товара": category,
                "Цвет": get_characteristic()["цвет"],
                "Материал": get_characteristic()["материал"],
                "Бренд": get_characteristic()["бренд"],
                "Отзыв": review
            })

print(f"✅ Сгенерировано {len(single_results)} отзывов по одной фразе")



# --------------------------
# Генерация отзывов с двумя фразами (по code)
# --------------------------

pair_results = []

# Лимиты пар
pair_limits = {
    102: 600,
    800: 100,
    801: 600
}

# Группируем фразы по code
code_to_phrases = defaultdict(list)
for group in key_words_all:
    code = group["code"]
    code_to_phrases[code].extend(group["phrases"])

# Убираем дубликаты
for code in code_to_phrases:
    code_to_phrases[code] = list(set(code_to_phrases[code]))

# Генерируем парные отзывы
for code, all_phrases in code_to_phrases.items():
    total_phrases = len(all_phrases)

    # Рассчитываем количество пар
    total_pairs = total_phrases * (total_phrases - 1) // 2
    max_by_ratio = round(total_pairs * 2 / 3)
    max_allowed = pair_limits.get(code, 100)
    num_pairs = min(max_by_ratio, max_allowed)

    print(f"Код {code}, всего фраз: {total_phrases} → будем генерировать пар: {num_pairs}")

    generated_pairs = set()

    while len(generated_pairs) < num_pairs and len(generated_pairs) < total_pairs:
        phrase1 = random.choice(all_phrases)
        phrase2 = random.choice([p for p in all_phrases if p != phrase1])

        pair = tuple(sorted((phrase1, phrase2)))

        if pair not in generated_pairs:
            generated_pairs.add(pair)

            product, category = get_random_product()
            review = generate_review_double(phrase1, phrase2, product)

            pair_results.append({
                "Тип отзыва": "Две фразы",
                "Код": code,
                "Фраза 1": phrase1,
                "Фраза 2": phrase2,
                "Товар": product,
                "Категория товара": category,
                "Цвет": get_characteristic()["цвет"],
                "Материал": get_characteristic()["материал"],
                "Бренд": get_characteristic()["бренд"],
                "Отзыв": review
            })

print(f"✅ Сгенерировано {len(pair_results)} отзывов по двум фразам")



# --------------------------
# Объединение результатов
# --------------------------

all_reviews = single_results + pair_results

# Сохраняем в Excel
df = pd.DataFrame(all_reviews)
df.to_excel("улучшенные_негативные_отзывы.xlsx", index=False)

print(f"\n✅ Всего сгенерировано: {len(df)} отзывов")
print("📌 Результат сохранён в файл: 'улучшенные_негативные_отзывы.xlsx'")
```
