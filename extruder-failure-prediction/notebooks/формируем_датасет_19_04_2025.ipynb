Создаём полный временной индекс с шагом 1 секунда и заполняем недостающие значения линейной интерполяцией

```python
# Убедимся, что колонка 'source_time' — это datetime
df_14['source_time'] = pd.to_datetime(df_14['source_time'])

# Сохраняем оригинальный DataFrame
df_14_original = df_14.copy()

# Делаем datetime индексом
df_14 = df_14.set_index('source_time')

# Создаем полный временной индекс с шагом 1 секунда
full_index = pd.date_range(start=df_14.index.min(), end=df_14.index.max(), freq='s')

# Восстанавливаем пропущенные секунды
df_14 = df_14.reindex(full_index)

# Интерполируем значения
df_14['value'] = df_14['value'].interpolate(method='linear')

# Возвращаем индекс обратно в колонку
df_14 = df_14.reset_index().rename(columns={'index': 'source_time'})


num_rows = len(df_14.index)
print(f"Количество строк в датасете: {num_rows:,}")

# Проверим результат
print(df_14.head(10))
```

```python
# Сохраняем полученный датасет
df_14.to_parquet("df_14_make_2.parquet", index=False)
```

Не удаляем выбросы по методу IQR

Аннотируем датасет df_14 добавлением mode
- mode = 1 если значение value != 0

```python
# Убедимся, что дата отсортирована по времени
df_14 = df_14.sort_values('source_time').reset_index(drop=True)

# Установим временной индекс для удобства
df_14.set_index('source_time', inplace=True)

# Создаем флаг нуля
df_14['is_zero'] = (df_14['value'] == 0).astype(int)

# Группируем подряд идущие одинаковые значения is_zero
df_14['group'] = (df_14['is_zero'] != df_14['is_zero'].shift()).cumsum()

# Считаем длину каждой группы
group_sizes = df_14.groupby('group').size()

# Отметим те группы, где длина >= 6 часов (6 * 60 * 60 = 21600 секунд)
long_zeros = group_sizes[group_sizes >= 21600].index

# Теперь создаем колонку флага активности
df_14['mode'] = 1
df_14.loc[df_14['group'].isin(long_zeros), 'mode'] = 0

# Удалим вспомогательные колонки
df_14 = df_14.drop(columns=['is_zero', 'group'])

# Вернем индекс обратно в колонку, если нужно
df_14 = df_14.reset_index()

# Проверим
print(df_14[['source_time', 'value', 'mode']].head(20))
```

```python
# Сохраняем полученный датасет
df_14.to_parquet("df_14_make_3.parquet", index=False)
```

# Дополняем датасет df_union вычисляемыми параметрами

```python
# Функция добавления в датасет вычисляемых параметров для заданного параметра датасета
# ВНИМАНИЕ!  num_lag должен быть чётным числом
def calculate_features(df, value_col, num_lag):

    if num_lag % 2 != 0:
        raise ValueError("num_lag должен быть четным числом")

    # Предшествующие значения до num_lag
    for i in range(1, num_lag + 1):  # Добавил +1, чтобы был лаг num_lag
        df[f'{value_col}_lag_{i}'] = df[value_col].shift(i)

    # Разность текущего и предшествующего значения на num_lag
    df[f'{value_col}_diff_1'] = df[value_col].diff(num_lag)

    half_num_lag = num_lag // 2  # Целочисленное деление

    # Среднее из предшествующих значений
    df[f'{value_col}_rolling_mean_{half_num_lag}'] = df[value_col].rolling(window=half_num_lag).mean()
    df[f'{value_col}_rolling_mean_{num_lag}'] = df[value_col].rolling(window=num_lag).mean()

    # Коэффициент автокорреляции
    df[f'{value_col}_autocorr_{half_num_lag}'] = df[value_col].rolling(window=half_num_lag).apply(lambda x: pd.Series(x).autocorr(), raw=False)
    df[f'{value_col}_autocorr_{num_lag}'] = df[value_col].rolling(window=num_lag).apply(lambda x: pd.Series(x).autocorr(), raw=False)

    # Дисперсия и стандартное отклонение
    df[f'{value_col}_rolling_var_{half_num_lag}'] = df[value_col].rolling(window=half_num_lag).var()
    df[f'{value_col}_rolling_var_{num_lag}'] = df[value_col].rolling(window=num_lag).var()
    df[f'{value_col}_rolling_std_{half_num_lag}'] = df[value_col].rolling(window=half_num_lag).std()
    df[f'{value_col}_rolling_std_{num_lag}'] = df[value_col].rolling(window=num_lag).std()

    return df
```
